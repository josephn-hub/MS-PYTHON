```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
plt.plot(x,y)

plt.xlabel("x axis")
plt.ylabel(" y axis")
plt.title("Line plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter plot graph")
plt.show()
```

```python
degree = ['Btech', 'Bsc', 'Pilot','Fashion']
student_count = [230, 130, 50, 70]

plt.bar(degree,student_count)
plt.xlabel("Degree")
plt.ylabel("Studentcount")
plt.show()
```

```python
degree = ['Btech', 'Bsc', 'Pilot','Fashion']
student_count = [230, 130, 50, 70]

plt.pie(student_count, labels=degree, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 =[1,3,5,7,9]

plt.subplot(3,1,1) # rows, column, plotnumber
plt.plot(x,y1)
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.title("Subplot1")

plt.subplot(3,1,2)
plt.scatter(x,y2)
plt.xlabel('X axis')
plt.ylabel('Y-axis')
plt.title("Subplot2")

#subplot3
plt.subplot(3,1,3)
plt.bar(y1,y2)
plt.xlabel("even")
plt.ylabel("odd")
plt.title("Subplot3")


plt.tight_layout()

plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import  load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X = iris.data
y = iris.target
```

```python
plt.scatter(X[y==0,0], X[y==0,1] , label='Setosa', c= 'red')
plt.scatter(X[y==1,0], X[y==1,1] , label='Versicolor', c= 'blue')
plt.scatter(X[y==2,0], X[y==2,1] , label='Verginica', c= 'green')
plt.legend()
plt.show()
```

```python
plt.hist(X[:,2], bins=12, edgecolor='black')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```
