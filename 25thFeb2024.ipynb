```python
import pandas as pd
import numpy as np
```

```python
d = {
    'team': ['SA','Newzealand','Australia','India','SA','Newzealand','Australia','India','SA','Newzealand','Australia','India'],
    'run': [30,40,50,60,20,70,80,80,35,50,90,10],
    'year': ['2010','2010','2012','2012','2013','2012','2010','2013','2010','2012','2013','2010']
}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
g= df.groupby('team')
g
```

```python
for team,run in g:
    print(team)
    print('-'*10)
    print(run)
```

```python
g.get_group('SA')
```

```python
g.min()
```

```python
g.max()
```

```python
g.mean()
```

```python
g_yr = df.groupby('year')
g_yr['run'].agg([np.min, np.max,np.mean, np.sum, np.std])
```

```python
Newyork = {
    'date':['01-02-12', '02-02-12','03-02-12','04-02-12'],
    'event':['hot','sunny','rain','cold'],
    'wind_speed':[33,23,45,24],
    'temp':[20,21,15,18]
}
```

```python
Paris={
    'date':['01-02-12', '02-02-12','03-02-12','04-02-12'],
    'event':['sunny','hot','cold','rain'],
    'wind_speed':[35,38,18,29],
    'temp':[30,40,15,20]
}
```

```python
nw = pd.DataFrame(Newyork)
nw
```

```python
ps = pd.DataFrame(Paris)
ps
```

```python
new_df = pd.concat([nw,ps])
new_df
```

```python
new_df = pd.concat([nw,ps], keys = ['Newyork', 'Paris'])
new_df
```

```python
new_df = pd.concat([nw,ps], ignore_index=True)
new_df
```

```python
temp1={
     'date':['01-02-12', '02-02-12','03-02-12','04-02-12'],
    'event':['sunny','hot','cold','rain'],
    'temp':[30,40,15,20]
}

temp_a = pd.DataFrame(temp1)
temp_a
```
