```python
# map

numbers = [1,2,3,4,5,6]
results_squares = list(map(lambda x :x**2, numbers))
print(type(results_squares))
```

```python
# filter

numbers = [1,2,3,4,5,6]
filter_even_result = list(filter(lambda x : x%2==0, numbers))
print(filter_even_result)
```

```python
# filter odd
numbers = [1,2,3,4,5,6]
filter_odd_result = list(filter(lambda x : x%2!=0, numbers))
print(filter_odd_result)
```

```python
import numpy as np

numbers = [1,2,3,4,5,6]
results_squares = np.array(list(map(lambda x :x**2, numbers)))
print(type(results_squares))
```
